### Spliting the dataset as independent variable X and dependent variable y

```python
X = df.iloc[:,:-1].values
y = df.iloc[:,-1].values
```

```python
print(X)
```

```python
print(y)
```

```python
X.shape
```

```python
y.shape
```

#### Spliting the X and y variables into the train and test dataset

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
X_test.shape
```

```python
y_test.shape
```

### Creating a Linear Regression Machine Learning Model

```python
from sklearn.linear_model import LinearRegression

linear = LinearRegression()
linear.fit(X_train,y_train)
```

```python
linear.coef_
```

```python
linear.intercept_
```

```python
y_pred_train = linear.predict(X_train)
y_pred_test = linear.predict(X_test)
```

### Comparing the Actual Values of the dataset with the Predicted Values of the Linear Model on to the test dataset

```python
df = pd.DataFrame({'Actual Values': y_test, 'Predicted Values': y_pred_test})  
df
```

```python
plt.scatter(y_train,y_pred_train)
```

```python
plt.scatter(y_test,y_pred_test)
```

```python
sns.distplot((y_test),bins=50)
```

```python
sns.distplot((y_pred_test),bins=50)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_absolute_percentage_error, mean_squared_error
M_A_E = mean_absolute_error(y_test, y_pred_test)
M_S_E = mean_squared_error(y_test, y_pred_test)
R_M_S_E = np.sqrt(mean_squared_error(y_test, y_pred_test))
M_A_P_E = mean_absolute_percentage_error(y_test, y_pred_test)
R2_score = r2_score(y_test, y_pred_test)
n = X_test.shape[0]
p = X_test.shape[1]
num = (1-R2_score)*(n-1)
den = n-p-1

adj_r2 = 1- (num/den)
```

```python
print("Mean Absolute Error : ", M_A_E)
print("Mean Square Error : ", M_S_E)
print("Root Mean Square Eror : ", R_M_S_E)
print("Mean Absolute Percentage Error : ", M_A_P_E)
print("R2 Score : ", R2_score)
print("Adjusted R2 Score : ", adj_r2)
```
